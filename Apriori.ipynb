```python
data=list(df["products"].apply(lambda x:x.split(',')))
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te_data = te.fit(data).transform(data)
df=pd.DataFrame(te_data,columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
df1 = apriori(df,min_support=0.2,use_colnames=True)
```

```python
df1
```

```python
df1.sort_values(by="support",ascending=False)
```

```python
df1['length'] = df1['itemsets'].apply(lambda x:len(x))
df1
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(df1,metric = "confidence", min_threshold=0.2)
```

```python
association_rules(df1,metric = "support", min_threshold=0.2)
```

```python
association_rules(df1,metric = "lift", min_threshold=0.2)
```
